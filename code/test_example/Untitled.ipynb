```python
import numpy as np
import time
from matplotlib import pyplot as plt 
from collections import defaultdict

def arr_lookup(n, arr):
    t0 = time.clock()
    top = arr.shape[0]
    for i in range(n):
        idx = np.random.randint(0,top-1)
        arr[idx]
    arr_add(arr)
    return(time.clock()-t0)

def dict_lookup(n,dict):
    t0 = time.clock()
    top = len(list(dict.keys()))
    for i in range(n):
        idx = np.random.randint(0,top-1)
        dict[idx]
    dict_add(dict)
    return(time.clock()-t0)

def dict_add(dict):
    for k in dict.keys():
        dict[k] + dict[k]

        
def arr_add(arr):
    arr + arr

```

```python
ns = np.arange(100, 50000, 100)
arr = np.zeros(100)
dict = {x:x for x in range(100)}
arr_times = []
dict_times = []
for n in ns:
    arr_times.append(arr_lookup(n, arr))
    dict_times.append(dict_lookup(n, dict))


    


```

```python
arr_line, = plt.plot(ns, arr_times, label = "arr")
dict_line, = plt.plot(ns, dict_times, label= "dict")

plt.legend(handles = [arr_line, dict_line])
```
